### Get the Pokemon Data from JSON

```python
#driver_standings_data = response.json()['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings']

#Pokemon Name Data
pok_name_data =response.json()['forms'][0]['name']
print(pok_name_data)
```

```python
#Pokemon Ability Data

pok_ability_data  = response.json()['abilities'][0]['ability']['name']
print(pok_ability_data)
```

```python
#Pokemon Base Experience Data

pok_baseexp_data = response.json()['base_experience']
print(pok_baseexp_data)
```

```python
#Pokemon URL for Sprite Data

pok_spriteurl_data = response.json()['sprites']['front_shiny']
print(pok_spriteurl_data)
```

```python
def try_pokemon(pok_name):
    print(pok_name)
    url = f'https://pokeapi.co/api/v2/pokemon/{pok_name}'
    response = requests.get(url)
    if response.status_code == 200:
        poke_data = response.json()
        my_pokedata = {
            'pk_name': poke_data['forms'][0]['name'], 
            'pk_ability': poke_data['abilities'][0]['ability']['name'],
            'pk_base_exp': poke_data['base_experience'],
            'pk_sprite_url': poke_data['sprites']['front_shiny']
        }
        return my_pokedata
    else:
        'That Pokemon does not exist.'


print(try_pokemon('pikachu'))
```
